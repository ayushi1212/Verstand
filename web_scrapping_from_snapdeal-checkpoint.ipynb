```python
my_url = 'https://www.snapdeal.com/search?keyword=sexual%20dildos&santizedKeyword=&catId=&categoryId=0&suggested=true&vertical=p&noOfResults=20&searchState=&clickSrc=suggested&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=ALL&url=&utmContent=&dealDetail=&sort=rlvncy'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
containers = page_soup.findAll("div",{"class": "col-xs-6" })
container = containers[0]
title = container.findAll("div",{"class":"product-tuple-image"})
#print(container.img["title"])
price = container.findAll("div",{"class":"product-tuple-description"})
#print(price[0].text)
filename = "products.csv"
f = open(filename, "w")
headers = "Product_Name, Pricing\n"
f.write(headers)

for container in containers:
    title_container = container.findAll("div",{"class":"product-tuple-image"})
    title = title_container[0].img["title"]
    price_container = container.findAll("div",{"class":"product-tuple-description"})
    price = price_container[0].text.strip()
    print("tile:" + title)
    print("price:" + price)
```
